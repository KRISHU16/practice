```python
# LIST AND TUPLES--------------------------------
# LIST[]---------
# List can store multiple data types and it is mutable we can change anything in a list 

A = [] # IT IS A EMPTY LIST

''' ---LIST INDEXING--- '''

friend = ["Apple","Ansh","Manish","Ram","Shubham",27,326.45,False,]
print(friend[0])
# OUTPUT = "Apple"

# JUST LIKE STRING INDEXING WORKS THE SAME IN LIST 

# variable[start:end:steps] 
# defult value of start = 0 , end = n , steps = 1
print(friend[1:4])
# OUTPUT = ['Ansh', 'Manish', 'Ram']
print(friend[::]) # IT PRINTS FROM (START TO END)
print(friend[::-1]) # IT WILL PRINT THE RESULT FROM (END TO START)||REVERSE

# LIST METHODS---
l1 = [1,8,7,2,21,15]
l2 = [1,8,7,2,21,15]
l3 = [1,8,7,2,21,15]
l4 = [1,8,7,2,21,15]
l5 = [1,8,7,2,21,15]
l6 = [1,8,7,2,21,15]

l1.sort() # REARRANGE THE LIST IN ASCENDING ORDER [1,2,7,8,15,21]
l2.reverse() # REVERCES THE LIST [15,21,2,7,8,1]
l3.append(8) # ADD 8 TO THE END OF THE LIST
l4.insert(3,8) # THIS WILL ADD 8 AT 3 INDEX
l5.pop(2) # WILL DELETE THE ELEMENT AT INDEX 2 AND RETURN ITS VALUE 
l6.remove(21) # WILL REMOVE 21 FROM THE LIST

print(l1)
print(l2)
print(l3)
print(l4)
print(l5)
print(l6)

# Because list is mutable if we change a list then the original list will change 
# unlike string which makes a copy of the multable string and it doesn't affect the original string

```

```python
# TUPLE--------------------

'''

-TUPLE IS IMMUTABLE UNLIKE LIST WHICH IS MUTABLE 
-IT IS DEFINED USGIN PARENTHESES ()
-AND LIST IS DEFIND USING BRACKET []

'''
t = () # IT IS A EMPTY TUPLE
t0 = (1) # IT IS NOT A TUPLE IT WILL IDENTIFY IT SELF AS A INTEGER 
t1 = (1,) # WE HAVE TO USE A (,) TO MAKE THIS VARIABLE A TUPLE OTHER WISE IT WILL IDENTIFY IT SELF AS INTEGER
print(t0,type(t0),t1,type(t1))

a = (1,3,234,524,False,"Rohan","Ram")
# TUPLE IS JUST LIKE LIST BUT IT IS IMMUTABLE WHICH MAKES IT DIFFRENT FROM LIST 

# METHODS OF TUPLE-------------

# 1. **`count(value)`** - Returns the number of times a specified value appears in the tuple.

my_tuple = (1, 2, 3, 1, 4, 1)
print(my_tuple.count(1))  # Output: 3

# 2. **`index(value)`** - Returns the index of the first occurrence of the specified value. Raises a `ValueError` if the value is not found.

my_tuple = (1, 2, 3, 4, 5)
print(my_tuple.index(3))  # Output: 2

# Tuple-related built-in functions:

# Even though tuples have few built-in methods, you can use several built-in functions with tuples:

# 3. **`len()`** - Returns the number of elements in the tuple.

my_tuple = (1, 2, 3, 4)
print(len(my_tuple))  # Output: 4


# 4. **`max()`** - Returns the largest element in the tuple (only works if the elements are comparable, like all numbers or all strings).

my_tuple = (10, 20, 30, 40)
print(max(my_tuple))  # Output: 40


# 5. **`min()`** - Returns the smallest element in the tuple.

my_tuple = (10, 20, 30, 40)
print(min(my_tuple))  # Output: 10


# 6. **`sum()`** - Returns the sum of all elements in the tuple (only works with numbers).

my_tuple = (10, 20, 30)
print(sum(my_tuple))  # Output: 60


# 7. **`sorted()`** - Returns a sorted list of the tuple's elements (does not modify the original tuple).

my_tuple = (3, 1, 4, 2)
print(sorted(my_tuple))  # Output: [1, 2, 3, 4]


# 8. **`in`** - Checks if an element exists in the tuple.

my_tuple = (1, 2, 3, 4)
print(2 in my_tuple)  # Output: True

# 9. **`tuple()`**- Converts another data type (such as a list) into a tuple.

my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)

my_tuple = (1,2,3,4,5)
repeting = my_tuple*3
print(repeting)
```
